## PCA Component

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
reduced=pd.DataFrame(pca.fit_transform(scaling))
reduced
```

## KMeans 

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=6,random_state=10)
kmeans.fit(reduced)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
y_pred=kmeans.predict(reduced)
y_pred
```

## Evaluate model

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(reduced,y_pred)
score
```

## Elbow Method

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(reduced)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,marker='o',color='b',linestyle='solid',markerfacecolor='red',markersize=10)
plt.title("Elbow Method")
plt.xlabel("No of Cluster")
plt.ylabel("WCSS")
plt.show()
```
